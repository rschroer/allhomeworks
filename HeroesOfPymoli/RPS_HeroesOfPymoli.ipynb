## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniq_items=purchase_data["Item ID"].nunique()
average_price=purchase_data["Price"].mean()
#It is at this point that I started to take objection to the structure of the product. I have observed that
# counting unique Item names will return 179 results, while Item IDs will return 183 items
# as there is a price difference in this, it makes me believe that instead of having some kind 
# sytem to handle discoutns for their items, that this company,
# literally creates new items in the same db with a new unique ID when something goes on sale.
# This will lead to future bugs such as if any of those items need to be adjusted,
# then multiple item IDS will need to have their properties altered
# I have decided to use Item Name as I feel this is a more accurate representation on the uniqueness of items
uniq_purchases=purchase_data["Purchase ID"].nunique()
total_revenue=purchase_data["Price"].sum()

Summary_Purchasing=pd.DataFrame({"Number of Unique Items":[uniq_items],
                                 "Average Price":[average_price],
                                 "Number of Purchases":[uniq_purchases],
                                 "Total Revenue":total_revenue})
#Format AVG Price And Total Revenue
Summary_Purchasing["Average Price"]=Summary_Purchasing["Average Price"].map( "${:.2f}".format)
Summary_Purchasing["Total Revenue"]=Summary_Purchasing["Total Revenue"].map( "${:.2f}".format)
Summary_Purchasing
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#remove duplicates who have done multiple purchases
unique_users=purchase_data.drop_duplicates("SN")
#count and create the summary dataframe
Gender_summary=unique_users.groupby("Gender").count()
#rename columns
Gender_summary=Gender_summary.rename(columns={"SN": "Total Count"})
#Drop other columns
Gender_summary=Gender_summary[["Total Count"]]
#sort the dataframe
Gender_summary=Gender_summary.sort_values("Total Count", ascending=False)
#remove index name
Gender_summary.index.names = ['']
#Calculate percenages
Gender_summary["Percentage of Players"] = 100*Gender_summary["Total Count"]/Gender_summary["Total Count"].sum()
#format percentage column
Gender_summary["Percentage of Players"]=Gender_summary["Percentage of Players"].map("{:.2f}".format)
Gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculate and create dataframe for purchase count
Purchase_Count=purchase_data[["Purchase ID","Gender"]].groupby("Gender").count()
#Calculate and create dataframe for mean price
Avg_price=purchase_data[["Price","Gender"]].groupby("Gender").mean()
#calculate and create dataframe for total 
tot_value=purchase_data[["Price","Gender"]].groupby("Gender").sum()
#find number of users
uniq_users=purchase_data[["SN","Gender"]].groupby("Gender").nunique()
#create seriese of avg/user
Avg_per_per=tot_value["Price"]/uniq_users["SN"]
#I like to merge it merge it
Purchase_Analysis=pd.merge(Purchase_Count, Avg_price, on="Gender")
Purchase_Analysis=pd.merge(Purchase_Analysis, tot_value, on="Gender")
Purchase_Analysis["Avg Total Purchase per Person"]=Avg_per_per
#rename all the columns
Purchase_Analysis.columns = ["Purchase Count", "Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]
#format all the numbers
Purchase_Analysis["Average Purchase Price"]=Purchase_Analysis["Average Purchase Price"].map( "${:.2f}".format)
Purchase_Analysis["Total Purchase Value"]=Purchase_Analysis["Total Purchase Value"].map( "${:.2f}".format)
Purchase_Analysis["Avg Total Purchase per Person"]=Purchase_Analysis["Avg Total Purchase per Person"].map( "${:.2f}".format)
Purchase_Analysis
# I thought about this later and I could have just done a column assignment instead of a merge. Can you speak to which is done faster?
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins
bins=[0,9,14,19,24,29,34,39,120]
#create labels
group_name=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
#Add column name to bin
purchase_data["Age Groups"]=pd.cut(purchase_data["Age"], bins, labels=group_name)
#It's at this time I realized I should have been dong nunique at the beginning of this exercise
demo_display=purchase_data.groupby("Age Groups").nunique()
demo_display.index.names = ['']
#This is to rename the column and drop the others without using the crazy typing and colons and stuff requred in the column rename syntax
demo_display["Total Count"]=demo_display["SN"]
#Calculate the percentages
demo_display["Percentage of Players"]= 100 * demo_display["Total Count"]/demo_display["Total Count"].sum()
#Add percentage column
demo_display=demo_display[["Total Count","Percentage of Players"]]
#Fix percentage formating
demo_display["Percentage of Players"]=demo_display["Percentage of Players"].map("{:.2f}".format)
demo_display
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Start making a new data frame
purchase_age=purchase_data.groupby("Age Groups").count()
#Add a new column as opposed to doing the rename column
purchase_age["Purchase Count"]=purchase_age["SN"]
#create a df to get the mean price
purchase_av_pr=purchase_data.groupby("Age Groups").mean()
#add that column to the first data frame
purchase_age["Average Purchase Price"]=purchase_av_pr["Price"]
#create a column to get the total purchase value
purchase_value=purchase_data.groupby("Age Groups").sum()
#add that column to the first data frame
purchase_age["Total Purchase Value"]=purchase_value["Price"]
#drop all the unceccesary columns
purchase_age=purchase_age[["Purchase Count","Average Purchase Price","Total Purchase Value"]]
#Create average column
purchase_age["Avg Total Purchase per Person"]=purchase_age["Total Purchase Value"] /purchase_age["Purchase Count"]
#Format columns
purchase_age["Average Purchase Price"]=purchase_age["Average Purchase Price"].map( "${:.2f}".format)
purchase_age["Total Purchase Value"]=purchase_age["Total Purchase Value"].map( "${:.2f}".format)
purchase_age["Avg Total Purchase per Person"]=purchase_age["Avg Total Purchase per Person"].map( "${:.2f}".format)
#remove the index namne
purchase_age.index.names = ['']
#Display results
purchase_age
#I kept this order because I feel it's better than the example order
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#staring group by
spender_groups=purchase_data.groupby("SN")
#find number of purchases
spender_df=spender_groups.count()
#create purchase count dolumn
spender_df["Purchase Count"]=spender_df["Purchase ID"]
#find avg price
spender_price=spender_groups.mean()
#create price column
spender_df["Average Purchase Price"]=spender_price["Price"]
#find total value
spender_value=spender_groups.sum()
#create value column
spender_df["Total Purchase Value"]=spender_value["Price"]
#sort by total price
spender_df=spender_df.sort_values(by=["Total Purchase Value"],ascending=False)
#get rid of index
spender_df.index.names = ['']
#format the prices
spender_df["Total Purchase Value"]=spender_df["Total Purchase Value"].map( "${:.2f}".format)
spender_df["Average Purchase Price"]=spender_df["Average Purchase Price"].map( "${:.2f}".format)
#keep the wanted columns
spender_df=spender_df[["Purchase Count","Average Purchase Price","Total Purchase Value"]]
#display summary
spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Initial DF creation
items_df=purchase_data[["Item Name","Item ID","Price"]]
#Initial Groupby
items_group=items_df.groupby(["Item ID","Item Name"])
#Create the summary dataframe and the purchase count
items_summary=items_group.count()
#Determine the item price. As they're all the same per Item ID, can use nmean here
price_group=items_group.mean()
#Add item price to summary dataframe
items_summary["Item Price"]=price_group["Price"]
#find out total price
price_sum=items_group.sum()
#add price to dataframe
items_summary["Total Purchase Value"]=price_sum["Price"]
#rename the price column to purchase count
items_summary=items_summary.rename(columns={"Price": "Purchase Count"})
#Copy this data frame to use in the next cell
items_summary2=items_summary.copy()
#sort by purchase count
items_summary=items_summary.sort_values(by=["Purchase Count"],ascending=False)
#format values
items_summary["Item Price"]=items_summary["Item Price"].map( "${:.2f}".format)
items_summary["Total Purchase Value"]=items_summary["Total Purchase Value"].map( "${:.2f}".format)
#display top 5
items_summary.head()
```
