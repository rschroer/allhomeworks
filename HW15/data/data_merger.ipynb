```python
import pandas as pd
import numpy as np
import os
```

# First, we're giong to check all the zip files for data errors

```python

for filename in os.listdir('zipped'):
    if filename.endswith(".zip"):
        file_path=os.path.join("zipped",filename)
        print(f"File: {file_path}")
        df=pd.read_csv(file_path,compression="zip")
        df=df.rename(str.lower, axis='columns')
        df.info()
        
df.columns
```

# Now, we're going to merge them and make one combined_bike_data.csv

```python
combined=pd.DataFrame(columns=['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender'])


for filename in os.listdir('zipped'):
    if filename.endswith(".zip"):
        file_path=os.path.join("zipped",filename)
        print(f"File: {file_path}")
        df=pd.read_csv(file_path,compression="zip")
        df=df.rename(str.lower, axis='columns')
        combined=combined.append(df,ignore_index=True)
        combined.info()
```
